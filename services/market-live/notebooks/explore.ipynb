# Explore market-live HTTP API

Call **market-live** for **current/latest** EOD-style bar reads (Tiingo-backed).

**Prerequisites**
- Run market-live locally (Compose uses port **8001**) or set `MARKET_LIVE_BASE_URL`.
- Venv: `make venv-service SERVICE=market-live` from repo root; `pip install jupyterlab ipykernel httpx` if needed.
- Real provider calls need `TIINGO_API_TOKEN` on the **server** process.

**Endpoints:** `GET /health`, `GET /current/{symbol}`.

Add more `.ipynb` files in this directory for ad-hoc experiments; keep `explore.ipynb` as the shared template.

```python
import json
import os

import httpx

BASE_URL = os.environ.get("MARKET_LIVE_BASE_URL", "http://localhost:8001").rstrip("/")
print("BASE_URL =", BASE_URL)
```

```python
with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/health")
    r.raise_for_status()
    print(json.dumps(r.json(), indent=2))
```

```python
symbol = "AAPL"

with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/current/{symbol}")
    if r.status_code == 404:
        print("404: no current bar (token missing, market closed, or symbol unknown)")
        print(r.text)
    else:
        r.raise_for_status()
        print(json.dumps(r.json(), indent=2, default=str))
```
