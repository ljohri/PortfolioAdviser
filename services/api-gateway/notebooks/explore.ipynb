# Explore api-gateway HTTP API

Call the **public SaaS facade** (routes to datalake + market-live upstream).

**Prerequisites**
- Run api-gateway (Compose uses port **8080**) and upstream `datalake` + `market-live`.
- Venv: `make venv-service SERVICE=api-gateway`; `pip install jupyterlab ipykernel httpx` if needed.

**Endpoints:** `GET /health`, `GET /v1/tickers` (see `app/routes/v1.py`).

Add more `.ipynb` files in this directory for ad-hoc experiments; keep `explore.ipynb` as the shared template.

```python
import json
import os

import httpx

BASE_URL = os.environ.get("API_GATEWAY_BASE_URL", "http://localhost:8080").rstrip("/")
print("BASE_URL =", BASE_URL)
```

```python
with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/health")
    r.raise_for_status()
    print(json.dumps(r.json(), indent=2))
```

```python
with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/v1/tickers")
    r.raise_for_status()
    data = r.json()
    print(f"tickers count: {len(data)}")
    print(json.dumps(data[:5], indent=2, default=str) if data else "(empty)")
```
