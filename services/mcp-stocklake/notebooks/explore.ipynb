# Explore APIs related to mcp-stocklake

**mcp-stocklake** is an **MCP (stdio) server**, not an HTTP service. OpenClaw talks to it over the MCP protocol.

To explore the **same canonical data and operations** from a notebook, call the **datalake HTTP API** that backs historical workflows (tickers, bars, backfill, etc.).

**Prerequisites**
- Run **datalake** (e.g. `http://localhost:8000`).
- Venv: `make venv-service SERVICE=mcp-stocklake` (includes datalake for local imports); `pip install jupyterlab ipykernel httpx` if needed.

**In-process alternative (advanced):** import `mcp_stocklake` / datalake runtime helpers from a Python session with `PYTHONPATH` or the service venv; keep HTTP below for a stable, service-boundary-friendly path.

Add more `.ipynb` files in this directory for ad-hoc experiments; keep `explore.ipynb` as the shared template.

```python
import json
import os

import httpx

BASE_URL = os.environ.get("DATALAKE_BASE_URL", "http://localhost:8000").rstrip("/")
print("datalake BASE_URL =", BASE_URL)
```

```python
with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/health")
    r.raise_for_status()
    print(json.dumps(r.json(), indent=2))
```

```python
with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/tickers")
    r.raise_for_status()
    tickers = r.json()
    print(f"tickers count: {len(tickers)}")
    print(json.dumps(tickers[:5], indent=2) if tickers else "(empty)")
```
