# Explore APIs related to mcp-market-live

**mcp-market-live** is an **MCP (stdio) server**. It proxies current-bar reads to **market-live** over HTTP.

This notebook calls **market-live** directly so you can probe the same payloads `get_current_bar` relies on.

**Prerequisites**
- Run **market-live** (e.g. `http://localhost:8001`).
- Venv: `make venv-service SERVICE=mcp-market-live`; `pip install jupyterlab ipykernel httpx` if needed.
- Server needs `TIINGO_API_TOKEN` for live Tiingo calls.

Add more `.ipynb` files in this directory for ad-hoc experiments; keep `explore.ipynb` as the shared template.

```python
import json
import os

import httpx

BASE_URL = os.environ.get("MARKET_LIVE_BASE_URL", "http://localhost:8001").rstrip("/")
print("market-live BASE_URL =", BASE_URL)
```

```python
with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/health")
    r.raise_for_status()
    print(json.dumps(r.json(), indent=2))
```

```python
symbol = "AAPL"

with httpx.Client(timeout=30.0) as client:
    r = client.get(f"{BASE_URL}/current/{symbol}")
    if r.status_code == 404:
        print("404: no current bar")
        print(r.text)
    else:
        r.raise_for_status()
        print(json.dumps(r.json(), indent=2, default=str))
```
